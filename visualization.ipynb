```python
import matplotlib.pyplot as plt
import pandas as pd

zadatak1 = pd.read_csv("./zadatak1_stats_history.csv")
zadatak2 = pd.read_csv("./zadatak2_stats_history.csv")
zadatak3 = pd.read_csv("./zadatak3_stats_history.csv")
zadatak4 = pd.read_csv("./zadatak4_stats_history.csv")
```

```python
zadatak1 = zadatak1.loc[zadatak1.Name != "Aggregated"].reset_index()
zadatak2 = zadatak2.loc[zadatak2.Name != "Aggregated"].reset_index()
zadatak3 = zadatak3.loc[zadatak3.Name != "Aggregated"].reset_index()
zadatak4 = zadatak4.loc[zadatak4.Name != "Aggregated"].reset_index()
```

```python
zadatak1
```

```python
zadatak1["id"] = zadatak1.index
zadatak2["id"] = zadatak2.index
zadatak3["id"] = zadatak3.index
zadatak4["id"] = zadatak4.index
```

```python
plt.plot(zadatak1["id"], zadatak1["Requests/s"])
plt.xlabel("Time")
plt.ylabel("Requests/s")
plt.title("Task 1 [local]")
plt.grid()
plt.show()
```

```python
plt.plot(zadatak2["id"], zadatak2["Requests/s"])
plt.xlabel("Time")
plt.ylabel("Requests/s")
plt.title("Task 2 [scaleset]")
plt.grid()
plt.show()
```

```python
plt.plot(zadatak3["id"], zadatak3["Requests/s"])
plt.xlabel("Time")
plt.ylabel("Requests/s")
plt.title("Task 3 [webapp]")
plt.grid()
plt.show()
```

```python
plt.plot(zadatak4["id"], zadatak4["Requests/s"])
plt.xlabel("Time")
plt.ylabel("Requests/s")
plt.title("Task 4 [function]")
plt.grid()
plt.show()
```

```python
plt.plot(zadatak1["id"], zadatak1["Requests/s"], label="Task 1 [local]")
plt.plot(zadatak2["id"], zadatak2["Requests/s"], label="Task 2 [scaleset]")
plt.plot(zadatak3["id"], zadatak3["Requests/s"], label="Task 3 [webapp]")
plt.plot(zadatak4["id"], zadatak4["Requests/s"], label="Task 4 [function]")
plt.xlabel("Time")
plt.ylabel("Requests/s")
plt.legend(loc='best')
plt.title("local vs scaleset vs webapp vs function")
plt.grid()
plt.show()
```
